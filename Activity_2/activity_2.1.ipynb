```python
# Data visuliztion 
import numpy as np 
from matplotlib import pyplot as plt 
```

```python
x=np.arange(1,10,0.1) # arange(LowerLimit,UpperLimit,StepValue)
y=2*x+5

# plt.plot(x,y)
# plt.show()

plt.plot(x,y,color='y') # Adding color to graph
plt.title('Line Plot',fontsize=20, color='green') # Adding title to graph
plt.xlabel('x-axis') # Adding x-axins name
plt.ylabel('y-axis') # Adding y-axis name
```

```python
# Line Graph
x = np.arange(1,10,0.1)
y1 = 2*x +5
y2 = 3*x + 5

plt.subplot(1,2,1)  # Plot subgraph subplot(region,parts,part)
plt.plot(x,y1)

plt.subplot(1,2,2)
plt.plot(x,y2)

plt.show()
```

```python
# Bar Graph
fruit = {'apple':30, 'mango':45, 'banana':10}
names = list(fruit.keys())
quantity = list(fruit.values())

plt.bar(names,quantity, color='r')
plt.title('Fruit Distribution')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
x = [10,20,30,40,50,60,70,80,90]
a = [1,2,3,4,5,4,3,2,9]
b = [5,6,7,8,3,2,1,4,9]

plt.figure(figsize=(10,5)) # increase FIgure Sixe
plt.scatter(x,a,s=200)
plt.scatter(x,b,s=500,marker='3')
plt.legend(['a','b']) # shows points/marker used for x and y
plt.title('X vs a&b')
plt.xlabel('X')
plt.ylabel('a&b')
```
